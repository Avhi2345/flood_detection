### Data Preprocessing

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
# Split the dataset
X = dummy_df.drop(columns=['FloodProbability'])
y = dummy_df['FloodProbability']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Scale the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

#### Importing the Requires Algorithms

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error, r2_score
```

### Model-1:- Linear Model

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

#### Predict and evaluate

```python
y_pred_lr=lr.predict(X_test)

# Evaluation Metrics
lr_mae = mean_absolute_error(y_test, y_pred_lr)
lr_rmse = np.sqrt(mean_squared_error(y_test, y_pred_lr))
lr_r2 = r2_score(y_test, y_pred_lr)
lr_mape = mean_absolute_percentage_error(y_test, y_pred_lr)
```

```python
print("Mean Absolute Error (MAE):", lr_mae)
print("Root Mean Squared Error (RMSE):", lr_rmse)
print("R² Score:", lr_r2)
print("Mean Absolute Percentage Error (MAPE):", lr_mape)
```

### Explanation:
1. The scatter plot should ideally have points close to the red line, which represents perfect prediction.
2. Deviations from the line indicate prediction errors, helping us assess model fit visually.


```python
# 1. Scatter Plot of Predicted vs. Actual
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_lr, alpha=0.6, color='b')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # Line for perfect prediction
plt.xlabel("Actual Flood Probability")
plt.ylabel("Predicted Flood Probability")
plt.title("Predicted vs Actual Flood Probability")
plt.show()
```

### Explanation:
1. A residual plot centered around zero with a bell-shaped curve suggests a good model fit.
2. Significant skewness may indicate areas where the model struggles with accuracy.


```python
# 2. Residual Plot
residuals_lr = y_test - y_pred_lr
plt.figure(figsize=(10, 6))
sns.histplot(residuals_lr, kde=True, color="purple")
plt.title("Residuals Distribution")
plt.xlabel("Residuals (Actual - Predicted)")
plt.ylabel("Frequency")
plt.show()
```

### Explanation:
1. A bar plot comparing different metrics (MAE, RMSE, R², MAPE) provides an at-a-glance view of model performance.
2. Lower MAE, RMSE, and MAPE indicate better accuracy, while a higher R² (close to 1) suggests strong predictive power.

```python
# 4. Error Metrics Comparison
metrics = ['MAE', 'RMSE', 'R²', 'MAPE']
values = [lr_mae, lr_rmse, lr_r2, lr_mape]

plt.figure(figsize=(10, 5))
sns.barplot(x=metrics, y=values, palette="coolwarm")
plt.title("Evaluation Metrics")
plt.ylabel("Score")
plt.show()
```

### Model-5:- Stacking Ensemble

```python
from sklearn.ensemble import StackingRegressor
```

```python
# Initialize base models
random_forest = RandomForestRegressor(n_estimators=100, random_state=42)
xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
lr = LinearRegression()

# Create the Stacking model with a Linear Regression meta-model
stacking_model = StackingRegressor(
    estimators=[
        ('rf', random_forest),
        ('xgb', xgb),
        ('lr', lr)
    ],
    final_estimator=LinearRegression()
)
```

```python
# Train the stacking model
stacking_model.fit(X_train, y_train)
```

### Predictions and evaluation

```python
y_pred_stack = stacking_model.predict(X_test)


# Evaluation Metrics
stack_mae = mean_absolute_error(y_test, y_pred_stack)
stack_rmse = np.sqrt(mean_squared_error(y_test, y_pred_stack))
stack_r2 = r2_score(y_test, y_pred_stack)
stack_mape = mean_absolute_percentage_error(y_test, y_pred_stack)
```

```python
print("Mean Absolute Error (MAE):", stack_mae)
print("Root Mean Squared Error (RMSE):", stack_rmse)
print("R² Score:", stack_r2)
print("Mean Absolute Percentage Error (MAPE):", stack_mape)
```

```python
# 1. Scatter Plot of Predicted vs. Actual
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_stack, alpha=0.6, color='pink')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # Line for perfect prediction
plt.xlabel("Actual Flood Probability")
plt.ylabel("Predicted Flood Probability")
plt.title("Predicted vs Actual Flood Probability")
plt.show()
```

```python
# 2. Residual Plot
residuals_stack = y_test - y_pred_stack
plt.figure(figsize=(10, 6))
sns.histplot(residuals_stack, kde=True, color="pink")
plt.title("Residuals Distribution")
plt.xlabel("Residuals (Actual - Predicted)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4. Error Metrics Comparison
metrics = ['MAE', 'RMSE', 'R²', 'MAPE']
values = [stack_mae, stack_rmse, stack_r2, stack_mape]

plt.figure(figsize=(10, 5))
sns.barplot(x=metrics, y=values, palette="coolwarm")
plt.title("Evaluation Metrics")
plt.ylabel("Score")
plt.show()
```
